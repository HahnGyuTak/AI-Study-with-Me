```python
from matplotlib.colors import hsv_to_rgb

V,H = np.mgrid[0:1:100j, 0:1:360j]
S = np.ones_like(V)

HSV_S100 = np.dstack((H,S, V))
RGB_S100 = hsv_to_rgb(HSV_S100)

HSV_S20 = np.dstack((H,S * 0.2, V))
RGB_S20 = hsv_to_rgb(HSV_S20)

HSV_S20.shape
```

```python
plt.subplot(211)
plt.imshow(RGB_S100, origin='lower', extent=[0,360,0,1], aspect=80)
plt.xlabel("색상(Hue)")
plt.ylabel("명도(Value)")
plt.title("채도 100")
plt.grid(False)

plt.subplot(212)
plt.imshow(RGB_S20, origin='lower', extent=[0,360,0,1], aspect=80)
plt.xlabel("색상(Hue)")
plt.ylabel("명도(Value)")
plt.title("채도 20")
plt.grid(False)

plt.tight_layout()
plt.show()
```

# Scikit-Image 이미지 처리

```python
import skimage
skimage.__version__
```

```python
import skimage

img_astro = skimage.data.astronaut()
img_astro.shape
```

```python
skimage.io.imsave("astronaut.png", img_astro)
img_astro2 = skimage.io.imread("astronaut.png")
```

이미지 읽고 쓰기

```python
from skimage import color

plt.subplot(131)
plt.imshow(img_astro)
plt.axis('off')
plt.title("RGB")

plt.subplot(132)
plt.imshow(color.rgb2gray(img_astro), cmap = plt.cm.gray)
plt.axis('off')
plt.title("Gray Scale")

plt.subplot(133)
plt.imshow(color.rgb2hsv(img_astro))
plt.axis('off')
plt.title("HSV")

plt.show()
```
